```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(0, 1.25, 500)
p = np.linspace(0, 2*np.pi, 500)
R, P = np.meshgrid(r, p)
Z = ((R**2 - 1)**2)

# Express the mesh in the cartesian system.
X, Y = R*np.cos(P), R*np.sin(P)

# Plot the surface.
ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0,alpha=1, edgecolor='none',antialiased=False)
#ax.plot_surface(X, Y, Z, facecolors=UWR(heatmap), linewidth=0.1,alpha=0.5, edgecolor='k',antialiased=True)



plt.show()
```
